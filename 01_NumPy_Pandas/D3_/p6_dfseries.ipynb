```python
import pandas as pd

s1=pd.Series(["A","B"],index=[1,2])
s2=pd.Series(["C","D"],index=[3,4])
s3=pd.Series(["E","F"],index=[5,6])

#默认按行拼接（竖着列下去）
pd.concat([s1,s2,s3])
```

```python
#按照索引对齐（没有索引的nan）
pd.concat([s1,s2,s3],axis=1)
```

```python
df1=pd.DataFrame(data={"a":[1,2],"b":[3,4]},index=[1,2])

s1=pd.Series(data=[7,10],index=[1,2],name="a")
pd.concat([df1,s1])
```

```python
pd.concat([df1,s1],axis=1)
```

```python
df1=pd.DataFrame(data={"a":[1,2],"b":[3,4]},index=[1,2])
df2=pd.DataFrame(data={"a":[7,8],"b":[9,10]},index=[1,2])
pd.concat([df1,df2],ignore_index=True)
```

```python
df1=pd.DataFrame(data={"a":[1,2],"b":[3,4]},index=[1,2])
df2=pd.DataFrame(data={"b":[7,8],"c":[9,10]},index=[2,3])
# join=outer 并集，保留所有行、列，没有的用nan填充

pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],join="inner")
```
