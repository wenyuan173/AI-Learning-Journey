```python
import pandas as pd
df = pd.DataFrame({"gmv":[100,200,300,400],"trade_date":["2024-01-06","2026-06-08","2025-11-23","2026-04-23"]})

# 新增一列dataframe，转化为日期类型
df["ymd"]=pd.to_datetime(df["trade_date"])
df
```

```python
df["yy"],df["mm"],df["dd"]= df["ymd"].dt.year,df["ymd"].dt.month,df["ymd"].dt.day
df
```

```python
df["week"]=df["ymd"].dt.day_name()
df
```
