```python
import numpy as np

arr1 =np.full((2,3),10)  #指定二维三列都是10
arr1
```

```python
arr2 = np.full_like(arr1,1)
arr2
```

```python
arr3= np.arange(0,10,2)  #从a到b每隔c个，包前不包后
arr3
```

```python
# 等差数列
# 分成c份 默认包含start和stop，如希望不包含，写endpoint=False
arr1= np. linspace(start=0,stop=10,num=7)
arr1
```

```python
a1= np.linspace(0,10,3, endpoint=False)
a1
```

```python
#等比数列
#2为底，在2-5中间平均取4个数做指数运算
a3 =np.logspace(start=2,stop=5,num=4,base=2)
a3
```
