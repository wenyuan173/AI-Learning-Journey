```python
import pandas as pd
from sqlalchemy import create_engine

df = pd.DataFrame({"age":[20,30,40,50],"name":["tom","alen","alex","emma"],"id":[101,102,103,104]})

USER='root'
PASSWORD='20021713'
HOST='localhost'
PORT='3306'
DB_NAME='AI2026'

engine = create_engine(
    f"mysql+pymysql://{USER}:{PASSWORD}@{HOST}:{PORT}/{DB_NAME}?charset=utf8mb4",
)

df.to_sql(name= "t_user",con = engine,if_exists="replace",index=False)

print ("save to database successfully")
```

```python
read_df= pd.read_sql_table("t_user",engine)
print(read_df)
```

```python
read_df1= pd.read_sql("select * from t_user where age >30",con=engine)
print(read_df1)
```
