```python
import numpy as np

arr1 =np.random.rand(2,3)
print(np.abs(arr1))   #绝对值
print(np.ceil(arr1))  #向上取整
print(np.floor(arr1))  #向下取整
print(np.rint(arr1))   #四舍五入
print(np.isnan(arr1))  #不是数值为true
print(np.multiply(arr1,2))  #乘法计算 每个元素*2
print(np.divide(arr1,arr1)) #除法 每个元素除以自己
print(np.where(arr1>0,1,0))  #元素大于0展示1，否则0
```
