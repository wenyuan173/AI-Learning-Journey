# 向前传播

## 单层单向RNN

```python
# 输入数据：(N,L,M) batch个数，句子长度，输入大小
input = torch.randn(2, 4,3)
```

```python
rnn1 = nn.RNN(3, 4, batch_first=True)
```

```python
output, hn = rnn1(input)
```

```python
print(output.shape)
print(hn.shape)
```

```python
print(output)
```

```python
print(hn)
```

## 双层单项

```python
rnn2 = nn.RNN(3, 4, num_layers=2, batch_first=True)
```

```python
output, hn = rnn2(input)
print(output.shape)
print(hn.shape)
print(output)
```

```python
print(hn)
```

## 单层双向RNN

```python
rnn3 = nn.RNN(3, 4,bidirectional=True, batch_first=True)
```

```python
output, hn = rnn3(input)
print(output.shape)
print(hn.shape)
print(output)
```

```python
print(hn)
```
