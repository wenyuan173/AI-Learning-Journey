## 简单语料测试

```python
from gensim.models import Word2Vec, KeyedVectors

# 准备语料数据
sentences = [['我','每天','乘坐','地铁','上班'],['我','每天','乘坐','公交','上班']]
```

```python
model = Word2Vec(
    sentences=sentences,
    vector_size=100, # 词向量维度
    window=5, # 上下文窗口大小
    min_count=1, # 最小词频（低于将忽略）
    sg =1, # 1是skip-gram，0是cbow
    workers=4
)
```

```python
print(model.wv['地铁'])
```

```python
print(model.wv.similarity('地铁','公交'))
```

## 基于真实语料训练模型

```python
import pandas as pd
import jieba
```

```python
# 加载数据
df = pd.read_csv('data/online_shopping_10_cats.csv').dropna()
print(df.head())
```

```python
# 分词，构建输入语料
sentences =[ [word for word in jieba.lcut(sentence) if word.strip() != ''] for sentence in df['review']]
```

```python
print(sentences[:5])
```

```python
print(len(sentences))
```

```python
# 定义并训练模型
model = Word2Vec(
    sentences=sentences,
    vector_size=100,
    window=5,
    min_count=2,
    sg =1,
    workers=4
)
```

```python
print(model.wv.similarity('公交','地铁'))
```

## 保存和加载训练好的词向量

```python
model_path = 'data/word2vec.model'
model.wv.save_word2vec_format(model_path)
```

```python
kv_model = KeyedVectors.load_word2vec_format(model_path)
```
