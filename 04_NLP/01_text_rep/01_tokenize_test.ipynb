```python
import jieba
```

```python
text = "扬哥毕业于德国慕尼黑大学统计与数学系"
```

## 精准模式

```python
generate = jieba.cut(text)
print(generate)
```

```python
for word in generate:
    print(word)
```

## 全模式

```python
generate = jieba.cut(text,cut_all=True)
for word in generate:
    print(word)
```

## 搜索引擎模式

```python
word_list = jieba.lcut_for_search(text)
print(word_list)
```

## 加入自定义词典

```python
text = "随着云计算技术的普及，越来越多企业开始采用云原生架构来部署服务，并借助大模型能力提升智能化水平，实现业务流程的自动化与智能决策。"
```

```python
word_list = jieba.lcut(text)
print(word_list)
```

```python
jieba.load_userdict("data/user_dict.txt")
```

```python
word_list = jieba.lcut(text)
print(word_list)
```

```python
# 只添加一个词
jieba.add_word("越来越多",20) # 后面的数字是词频
word_list = jieba.lcut(text)
print(word_list)
```
