```python
from gensim.models import KeyedVectors
```

```python
model_path = 'data/sgns.weibo.word'
model = KeyedVectors.load_word2vec_format(model_path)
```

```python
# 查看词向量长度
print(model.vector_size)
```

```python
print(len(model.key_to_index))
```

```python
# 查看词向量矩阵
print(model.vectors)
```

```python
# 查看某个词的词向量
id = model.key_to_index['地铁']
print(id)
```

```python
print(model['地铁'])
print(model['公交'])
```

## API测试

```python
print(model.similarity('地铁','公交')) # 计算相似度
```

```python
# 查询近义词
similar_words = model.similar_by_word('上班',topn=8)
print(similar_words)
```

```python
similar_words= model.most_similar(positive=['地铁','公交'])
print(similar_words)
```

```python
similar_words = model.most_similar(positive=['男人','女孩'],negative=['女人'])  # 对比找差：男人如果对应女孩，女人对应什么
print(similar_words)
```
