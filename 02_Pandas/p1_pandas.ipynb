```python
#通过列表创建Series
import pandas as pd

s1= pd.Series([1,2,3,4,5])
s1
```

```python
#指定series对象的名字和索引名称
s2= pd.Series([1,2,3,4,5],index=['a','b','c','d','e'],name='index')
s2
```

```python
dic_1={"a":1,"b":2,"c":3,"d":4,"e":5}
s3 = pd.Series(dic_1,index=['a','d','e'],name='index2')
s3
```

以上是创建Series的三种方式
