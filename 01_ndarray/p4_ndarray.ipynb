ndarray生成随机元素数组

```python
import numpy as np

arr1 = np.random.rand(2,3)
arr1
```

```python
arr2 = np.random.randint(0,10,(2,3)) #随机生成0-10之间的元素，2行3列
arr2
```

```python
arr3= np.random.uniform(0,10,(3,3))
arr3
```

```python
arr4 =np.random.randn(2,3)
arr4
```
