介绍ndarray的属性


```python
import numpy as np

arr1=np.array([[1,2,3],[4,5,6]])
print(arr1)
print(type(arr1))
print("维度",arr1.ndim)
print("shape",arr1.shape)
print("type",arr1.dtype)
print("size",arr1.itemsize)
```

```python
data =[1,2,3]
print(f"data列表的id为{id(data)}")
arr2=np.array(data)
print(f"arr2列表的id为{id(arr2)}")
arr3=np.array(data)
print(f"arr3列表的id为{id(arr3)}")
```

```python
arr4=np.array((1,2,3))
a1= np.array(arr4)
a2= np.asarray(arr4)
print(f"arr4的数组地址值为{id(arr4)}")
print(f"a1数组的地址值为{id(a1)}")
print(f"a2数组的地址值为{id(a2)}")
```
