```python
import torch
import torch.nn as nn
```

## 二元交叉熵损失函数

```python
# 1.定义5条数据
output = torch.randn(5)
# 2.使用sigmoid映射成概率
proba=torch.sigmoid(output)
# 3.定义目标值
target = torch.tensor([0,0,1,0,1]).float()
# 4.定义损失函数
loss_fn = nn.BCELoss()
# 5.计算损失
loss = loss_fn(proba, target)
print(loss)
```

```python
loss_fn = nn.BCEWithLogitsLoss() # 把传入的数据先sigmoid然后再使用BCE计算损失
loss = loss_fn(output, target)
print(loss)
```

## 交叉熵损失函数

```python
# 1.定义输出值
output = torch.randn(8,6)

# 2.定义损失函数  交叉熵损失+softmax
loss_fn = nn.CrossEntropyLoss()
```

```python
# 3.目标是标签编码
target = torch.tensor([0,0,1,0,1,5,1,3])

# 4.计算损失
loss = loss_fn(output, target)
print(loss)
```

```python
# 4.目标编码也支持独热编码：target应该是（8，6）的矩阵
target_onehot = torch.zeros(8,6)
target_onehot[torch.arange(8),target] = 1
print(target_onehot)
```
