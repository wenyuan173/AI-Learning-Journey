```python
import torch
from torch.utils.data import Dataset,TensorDataset
```

## 自定义数据集

```python
class MyDataset(Dataset):
    def __init__(self,data):# 初始化
        self.data = data
    def __len__(self): # 获取数据集大小
        return len(self.data)
    def __getitem__(self, idx): # 返回索引号单条数据（告诉外界第n条数据有什么）
        return self.data[idx]
```

```python
ds = MyDataset([10,20,30,40,50]) # 利用数据列表创建数据集

print(ds[0])
print(len(ds))
```

## 内置数据集（Tensor数据集）

```python
tds= TensorDataset(torch.randn(10))
print(tds[0])
print(len(tds))
```

## 数据加载器（DataLoader）

```python
from torch.utils.data import DataLoader
```

```python
x = torch.randn(10,3)
y = torch.randn(10)
tds = TensorDataset(x,y)

loader= DataLoader(tds,batch_size=3,shuffle=True, drop_last=False)
```

```python
for input,target in loader:
    print(input,target)
    print("--------------")
```
