```python
import torch
import torch.nn as nn
```

```python
# 定义目标值和预测值（nn的输出）
output = torch.randn(5)
target = torch.randn(5)
print(output)
print(target)
```

## MAE均绝对误差(L1 loss)

```python
loss_fn = nn.L1Loss()
loss= loss_fn(output, target)
print(loss)
```

## MSE均方误差

```python
loss_fn = nn.MSELoss()
loss= loss_fn(output, target)
print(loss)
```

## Smooth L1

```python
loss_fn = nn. SmoothL1Loss()
loss= loss_fn(output, target)
print(loss)
```
