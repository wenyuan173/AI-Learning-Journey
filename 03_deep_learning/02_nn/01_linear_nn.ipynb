```python
import torch
import torch.nn as nn
```

```python
linear = nn.Linear(in_features=2,out_features=3)
```

```python
print(linear.weight)
print(linear.bias)
```

## 向前传播

```python
input = torch.tensor([10,20]).float()
```

```python
out = linear.forward(input)
print(out)
```

```python
# 构造数据：多条
input = torch.tensor([[10,20],[100,200]]).float()
out = linear.forward(input)
print(out)
```
