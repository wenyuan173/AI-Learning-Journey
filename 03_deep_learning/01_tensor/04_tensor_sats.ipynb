```python
import torch
```

```python
t1 = torch.randint (1,10,(2,3,4)).float()
print(t1)
```

```python
print(t1.sum())
```

```python
t1.sum(dim=0)  #消维度：矩阵同位置相加
```

```python
print(t1.sum(dim=1)) #每个矩阵变成一行
```

```python
print(t1.sum(dim=2)) #每个矩阵横向相加
```

```python
print(t1.mean(dim=0))
```
