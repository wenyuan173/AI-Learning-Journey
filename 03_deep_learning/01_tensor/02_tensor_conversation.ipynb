```python
import torch
```

```python
t1 = torch.tensor([1, 2, 3])
print(t1.dtype)
```

```python
t2=t1.type(torch.float)
print(t2.dtype)
```

```python
t2 =t1.to(torch.float)
print(t2.dtype)
```

```python
t2 =t1.float()
print(t2.dtype)
```
