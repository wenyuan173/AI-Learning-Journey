```python
import torch
t= torch.randint(0,10,(2,5,4))
print(t)
```

## 1.简单索引

```python
print(t[0][0][0])
print(t[0,0,0])
```

```python
print(t[0])
```

```python
print(t[0,0])
```

## 2.范围索引

```python
print(t[1:])
```

```python
print(t[1,1:,2:])
```

## 3。列表索引,

```python
list1 =[0,1,1] # 定位第几个矩阵
list2 =[1,3,2] # 定位第几行
# 上下一一对应
print(t[list1,list2])
```

```python
list1 =[[0],[1]]
list2 =[1,3,2]
# 上面一个取下面所有
print(t[list1,list2])
```

## 4.布尔索引

```python
mask = t >5
mask
```

```python
print(t[mask])
```

```python
# 取出每行首字母大于5的行
# 先筛选
mask = t[:,:,0]>5
print(mask)
# 再取出
print(t[mask])
```

```python
# 取出每列的首元素大于5的那些列
mask = t[:,0,:]<5
print(mask)
```

```python
print(t.mT[mask].T)
```
