```python
import torch
import numpy as np
```

1.基于内容创建tensor

```python
t1= torch.tensor([10,20,30])
print(t1)
print(t1.shape)
```

```python
t1= torch.tensor([[10,20,30],[40,50,60]])
print(t1)
print(t1.shape)
```

```python
t1= torch.tensor([[[10,20,30],[40,50,60]],[[10,20,30],[40,50,60]]])
print(t1)
print(t1.shape)
```

```python
t1= torch.tensor(np.array([[10,20,30],[40,50,60]]))
print(t1)
print(t1.shape)
print(t1.dtype)
```

## 2.基于形状创建

```python
t1 =torch.Tensor(2,3,5)
print(t1)
print(t1.shape)
print(t1.dtype)
```

## 3.创建指定类型的张量

```python
t1 = torch.tensor([[10,20,30],[40,50,60]],dtype=torch.float32)

print(t1.dtype)
```

```python
t1 = torch.IntTensor(2,3)
print(t1.dtype)
```

```python
t1= torch.LongTensor(2,3)
print(t1.dtype)
```

```python
t1= torch.FloatTensor([10,20,30])
print(t1.dtype)
```

```python
t1= torch.HalfTensor([10,20,30])
print(t1.dtype)
```

 ## 4.指定区间创建

```python
t1 =torch.arange(10)
print(t1)

t1= torch.arange(start=0,end=10,step=2)
print(t1)
```

```python
t1= torch.linspace(1,10,3)
# 首和尾还有输出个数
print(t1)
```

```python
t1=torch.logspace(1,3,3)
print(t1)
```

## 5.根据填充创建

```python
t1= torch.ones(2,3)
print(t1)
```

```python
t1 = torch.zeros(2,3)
print(t1)
t2 = torch.zeros_like(t1)
print(t2)
```

```python
t1 = torch.full((2,3),5)
print(t1)
```

```python
t1 = torch.eye(5,4) #单位矩阵
print(t1)
```

## 6.随机创建张量

```python
t1= torch.rand(2,3)# 均匀分布：【0，1】
print(t1)
```

```python
t1 =torch.randint(low=10,high=100,size=(2,3)) # 均匀分布
print(t1)
```

```python
t1= torch.randn(2,3) # 标准正态分布
print(t1)
```

```python
t1 = torch.normal(mean= 5,std= 1,size= (2,3)) # 均值是5，标准差为1，二行三列的张量
print(t1)
```

```python
t1 = torch.randperm(10)  # 打乱
print(t1)
```
