## 2.元素级运算

```python
t1 = torch.tensor([[1,2,3],[4,5,6]])
t2 = torch.tensor([[7,8,9],[4,5,6]])
```

```python
print(t1+t2) # 相加，形状一样，或者满足广播条件
```

```python
t3 = torch.tensor([[9],[6]])
print(t1+t3)
```

```python
print(t1 * t2)  # 哈德马积
```

## 3.矩阵相乘

```python
t1 = torch.tensor([[1,2,3],[4,5,6]])
t2 = torch.tensor([[8,9],[4,5],[7,9]])

print(t1 @ t2)
print(t1.mm(t2)) # 必须是矩阵
```

```python
t1= torch.tensor([[[1,2,3],[4,5,6],[6,5,4]],[[3,2,1],[3,2,1],[3,2,1]]])
t2 =torch.tensor([[[1,2],[3,4],[5,6]],[[6,5],[4,3],[2,1]]])
print(t1.shape,t2.shape)
```
