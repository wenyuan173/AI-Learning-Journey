```python
import torch
t= torch.randint(0,10,(2,5,4))
print(t)
```

## 维度变换

```python
print(t.T)# (0,2,3) =>(3,2,0)
```

```python
print(t.mT) # 只转置最后两个维度
```

```python
print(t.transpose(0,1))
# 指定交换两个维度
```

```python
print(t.permute(0,2,1))
print(t.permute(2,0,1).shape)
```

## 形状调整

```python
t1= t.reshape(-1)
print(t1)
print(t1.shape)
```

## 增删维度

```python
t1 = torch.randint(1,10,(2,4))
print(t1)
print(t1.reshape(2,1,4))
```

```python
print (t1.unsqueeze(dim=0))
```

```python
t1 =torch.randint(1,10,(2,1,3))
t2 = torch.randint(1,10,(2,3,3))
print(t1)
print(t2)
```
